### Sección para recorrer cada página de FincaRaiz

```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from webdriver_manager.firefox import GeckoDriverManager

driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))

for x in range(1,10):

    driver.get('https://www.fincaraiz.com.co/finca-raiz/venta/medellin/antioquia/usados/pagina'+str(x))

    sleep(2)

    all_href = driver.find_elements(By.XPATH,  "//a[contains(@title, 'Venta')]")

    with open('paginas_medellin.csv', 'a', newline='') as file:
        writer = csv.writer(file)

        for my_href in all_href:

            writer.writerow([my_href.get_attribute("href"), x])
 
driver.close()
driver.quit()
```

### Sección para descarga de información indivual por predio

Sección para la descargar del contenido por página

```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from webdriver_manager.firefox import GeckoDriverManager
import json

driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))

with open('paginas_medellin.csv', 'r', newline='') as file:
    data = list(csv.reader(file))

for x in range(1,len(data)):
    
    driver.get(str(data[x][0]))
    print(data[x][0])
    

    sleep(3)

    precio=0
    descripcion=""
    imagen=""
    direccion=""
    lat=0
    lon=0

    try:
        precio = 0
    except NoSuchElementException:  #spelling error making this code not work as expected
        print(NoSuchElementException)
        pass
        
    try:
        info = driver.find_elements(By.XPATH,  "//script[@type='application/ld+json']")
        y = json.loads(info[0].get_attribute('innerHTML'))
        descripcion=y["description"].replace('\n', ' ')
        precio= y["priceSpecification"]["price"]
        imagen=y["image"]
        direccion=y["object"]["address"]
        lat=y["object"]["geo"]["latitude"]
        lon=y["object"]["geo"]["longitude"]
        text_box = driver.find_element(by=By.NAME, value="m²")
        print(text_box)
        
    except NoSuchElementException:  #spelling error making this code not work as expected
        print(NoSuchElementException)
        pass
    

    with open('data_medellin.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([data[x][0], descripcion,precio,imagen,direccion,lat,lon])

```
